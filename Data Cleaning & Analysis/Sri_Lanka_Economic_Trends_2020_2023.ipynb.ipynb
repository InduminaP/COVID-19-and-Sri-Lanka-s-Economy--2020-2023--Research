```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style for consistency
sns.set(style="whitegrid")

# Load the dataset
df = pd.read_csv("sri_lankan_economic_data_of_2020-2023_v01.csv")

# Display basic information about the dataset
df.info()
```

```python
# Load the dataset
# Transpose the DataFrame to switch rows and columns
df_transposed = df.transpose().reset_index()

# Print the first few rows of the transposed DataFrame to verify the transformation
print("\nTransposed DataFrame using transpose():")

# Replace the specific value "Series" with "Year" across the entire DataFrame
# This is useful for improving clarity and readability of the DataFrame
df_transposed.replace("Series", "Year", inplace=True)

# Display the first 10 rows of the modified transposed DataFrame
df_transposed.head(10)
```

```python
# Set the first row of the transposed DataFrame as the new header
# This replaces the default integer column names with meaningful labels from the first row
df_transposed.columns = df_transposed.iloc[0]

# Drop the first row from the DataFrame since it is now used as the header
# The reset_index(drop=True) method is used to reindex the DataFrame without adding the old index as a column
df_transposed = df_transposed.drop(0).reset_index(drop=True)

# Display the first few rows of the updated DataFrame to verify the changes
df_transposed.head()
```

```python
# Convert specified columns to float data type to ensure numerical operations can be performed accurately
df_transposed = df_transposed.astype({
    'Total Population (Millions)': float,
    'Population growth (annual %)': float,
    'GNI per Capita (US$, Atlas Method)': float,
    'GNI (PPP, Current International $,Billions)': float,
    'GNI per Capita (PPP, Current International $)': float,
    'Life Expectancy at Birth ': float,
    'Fertility Rate (Births per Woman)': float,
    'Primary Completion Rate (% of Age Group)': float,
    'Secondary School Enrollment (% Gross)': float,
    'Electric Power Consumption (kWh per Capita)': float,
    'GDP (Current US$,Billions)': float,
    'GDP Growth (Annual %)': float,
    'Inflation (GDP Deflator, Annual %)': float,
    'Agriculture, Forestry & Fishing (% of GDP)': float,
    'Industry (Including Construction) (% of GDP)': float,
    'Exports of Goods & Services (% of GDP)': float,
    'Imports of Goods & Services (% of GDP)': float,
    'Revenue Excluding Grants (% of GDP)': float,
    'Mobile Cellular Subscriptions (per 100 People)': float,
    'External Debt Stocks (Current US$,Billions)': float,
    'Total Debt Service (% of Exports)': float,
    'Personal Remittances Received (Current US$,Billions)': float,
    'Foreign Direct Investment Inflows (Current US$,Billions)': float,
    'Unemployment Rate (% of Labor Force)': float
})

# Convert the 'Year' column to datetime format to facilitate date manipulations
df_transposed['Year'] = pd.to_datetime(df_transposed['Year'])

# Extract only the year component from the datetime and update the 'Year' column
df_transposed['Year'] = df_transposed['Year'].dt.year

# Display information about the DataFrame to verify data types and non-null counts of columns
df_transposed.info()
```

```python
# Data cleaning and preprocessing
df_transposed.to_csv('cleaned_sri_lankan_economic_data_of_2020-2023_v02.csv', index=False)
```

# Basic visualizations for future reports

```python
plt.plot(df_transposed['Year'], df_transposed['Total Population (Millions)'], 
         label='Total Population', color='blue', marker='o')
plt.title('Total Population Over Years')
plt.xlabel('Year')
plt.ylabel('Total Population (Millions)')

# Set x-ticks to show only whole numbers for the years
plt.xticks(df_transposed['Year']) 

plt.grid(False)
plt.show()
```

```python
# Create the bar chart
plt.bar(df_transposed['Year'], df_transposed['Secondary School Enrollment (% Gross)'], color='green')
plt.title('Secondary School Enrollment')
plt.xlabel('Year')
plt.ylabel('School Enrollment (%)')
plt.xticks(df_transposed['Year'])
plt.grid(False)

# Add value labels on top of the bars
for i in range(len(df_transposed['Year'])):
    plt.text(x=df_transposed['Year'][i], 
             y=df_transposed['Secondary School Enrollment (% Gross)'][i] + 0.5,  # Adjust the offset as needed
             s=str(df_transposed['Secondary School Enrollment (% Gross)'][i]),
             ha='center')

# Show the plot
plt.show()
```

```python
years = df_transposed['Year']
gni_per_capita = df_transposed['GNI per Capita (US$, Atlas Method)']

# Plotting the area chart
plt.fill_between(years, gni_per_capita, color="skyblue", alpha=0.4)  # Shaded area
plt.plot(years, gni_per_capita, color="Slateblue", alpha=0.8, linewidth=2)  # Line on top

# Adding value labels
for i, value in enumerate(gni_per_capita):
    plt.text(years[i], value + 100, f'{value:.2f}', ha='center', va='bottom', fontsize=8)  # Adjust positioning

# Adding labels and title
plt.title("GNI per Capita Over Time")
plt.xlabel("Year")
plt.ylabel("GNI per Capita (US$)")
plt.xticks(df_transposed['Year'])
plt.grid(False)

# Display the plot
plt.show()
```
